Task 2 — Data Cleaning

```python
df = df.rename(columns={'listed_price': 'selling_price'})
df
```

```python
df = df.dropna(subset=['selling_price'])
df
```

```python
drop_cols = [
    'usedCarSkuId', 'images', 'top_features', 'comfort_features',
    'interior_features', 'exterior_features', 'safety_features'
]

df = df.drop(columns=drop_cols, errors='ignore')
df
```

```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

cat_cols = df.select_dtypes(include=['object']).columns
df[cat_cols] = df[cat_cols].fillna(df[cat_cols].mode().iloc[0])
df
```

```python
df['oem'] = df['oem'].str.strip().str.lower()
df
```

```python
print("Duplicates:", df.duplicated().sum())
df = df.drop_duplicates()
```

```python
df = df[df['selling_price'] > 0]
df
```

```python
print(df.shape)
df.head()
```

Task 3 — Baseline MAE

```python
mean_price = df['selling_price'].mean()
print("Mean Price:", mean_price)
```

```python
df['baseline_pred'] = mean_price
```

```python
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(df['selling_price'], df['baseline_pred'])

print("Baseline MAE:", mae)
```

The dataset does not contain a mileage column, so the mileage step was not applicable.
The dataset was cleaned by handling missing values, removing duplicates, and standardizing columns.
A baseline model was created using the mean selling price.
The Mean Absolute Error (MAE) obtained is 551116.1133220675.
This baseline serves as a reference for future machine learning models.
